```python
count=df.iloc[:,:].sum()
count.sort_values(0, ascending=False, inplace=True)
count=count.to_frame().reset_index()
count=count.rename(columns={'index':'Movie_names', 0:'Count'})
count
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,7))
ax= sns.barplot(x='Count', y='Movie_names', data=count)
plt.xticks(fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.ylabel('Frequency', fontsize=14)
plt.title('Moviess purchase Frequency', fontsize=16, fontweight='bold')

plt.show()
```

Association Rule

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

A. **10% Support 30% Confidence**

```python
freq_movies_1=apriori(df, min_support=0.1, use_colnames=True)
freq_movies_1
```

```python
rule1=association_rules(freq_movies_1, metric='confidence', min_threshold=0.3)
rule1
```

```python
rule1.sort_values('lift', ascending=False)
```

```python
lift_1= rule1[rule1['lift']>=1]
lift_1
```

```python
plt.figure(figsize=(15,7))
plt.scatter(lift_1['support'], lift_1['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(lift_1['support'], lift_1['lift'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

```python
matrix_1=lift_1.pivot('antecedents','consequents','lift')
matrix_1
```

```python
plt.figure(figsize=(20,8), dpi=250)

sns.heatmap(matrix_1, annot=True)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```
